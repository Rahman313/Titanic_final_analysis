```python
import seaborn as sns
sns.histplot(data=df,x='Sex')
```

```python
sns.histplot(data=df,y='SibSp')
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.displot(df['Age'])
```

```python
sns.histplot(df['Fare'], bins=40, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
genderwise_survival=df.groupby('Sex')['Survived'].value_counts().unstack()
genderwise_survival=genderwise_survival.plot(kind='bar')
plt.title("survivors of titanic")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.legend(['did not survived','Survived'])
plt.show()
df['Sex'].value_counts()
```

```python
classwise_survival=df.groupby('Pclass')['Survived'].value_counts().unstack()
classwise_survival=classwise_survival.plot(kind='bar')
plt.title("survivores by class")
plt.xlabel('class')
plt.ylabel('Survived')
plt.xticks(rotation=0)
plt.yticks(rotation=30)      #rotaional of axis values
plt.legend(['did not survived','Survived'])
plt.show()
df['Pclass'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```
